```python
#Import library
import pandas as pd
import numpy as np
```

```python
persona = pd.read_csv('persona.csv')
persona.head()
```

```python
# Analisis statistik deskriptif
persona.shape  # Mengembalikan dimensi DataFrame
persona.describe().T # Menghasilkan statistik deskriptif
```

```python
persona.isnull().values.any() # Returns any value is missing in DataFrame
persona.isnull().sum() #  Returns how many missing values exist in the DataFrame
```

```python
print("1.Ada berapa sumber unik dan Berapa frekuensi mereka?")
print("Unique SOURCE", persona["SOURCE"].nunique())
print("Frequencies of SOURCE:\n", persona['SOURCE'].value_counts())
print("-------------------------------------------------------------")
print("2.Ada berapa harga unik??")
print("Unique PRICE\n", persona["PRICE"].nunique())
print("-------------------------------------------------------------")
print("3.Berapa banyak penjualan yang dilakukan dari harga yang mana?")
print("Frequencies of PRICE\n", persona['PRICE'].value_counts())
print("-------------------------------------------------------------")
print("4.Berapa banyak penjualan yang dilakukan dari negara mana?")
print("Frequencies of COUNTRY\n", persona['COUNTRY'].value_counts())
print("-------------------------------------------------------------")
print("5.Berapa banyak yang diperoleh dari penjualan menurut negara?")
print("Winnings by COUNTRY:\n", persona.groupby("COUNTRY")[['PRICE']].aggregate("sum"))
print("-------------------------------------------------------------")
print("6.Berapa angka penjualan berdasarkan jenis sumber?")
print("Amount of sales by SOURCE\n", persona['SOURCE'].value_counts())
print("-------------------------------------------------------------")
```

```python
print("Rata-rata harga menurut negara")
print(persona.groupby('COUNTRY')[["PRICE"]].aggregate("mean"))
print("Rata-rata harga menurut sumber")
print(persona.groupby('SOURCE')[["PRICE"]].aggregate("mean"))
print("Negara dan Sumber rincian rata-rata pendapatan?")
persona.groupby(["COUNTRY", 'SOURCE'])["PRICE"].mean()
```

```python
# Pendapatan rata-rata berdasarkan variabel
agg_df = persona.groupby(["COUNTRY", 'SOURCE', "SEX", "AGE"])["PRICE"].mean().sort_values(ascending=False)
agg_df.head()
```

```python
#Ubah nama indeks menjadi nama variabel
agg_df = agg_df.reset_index() # Jika Salah, kembalikan salinannya. Jika tidak, lakukan operasi di tempat dan kembalikan Tidak Ada.
agg_df.head() # Nomor indeks baru
```

```python
# Ubah variabel AGE menjadi variabel kategori dan tambahkan ke to agg_df

my_labels = ['0_18', '19_23', '24_30', '31_40', '41_70']
agg_df["AGE_CUT"] = pd.cut(x=agg_df["AGE"], bins=[0, 18, 23, 30, 40, 70], labels=my_labels)
agg_df.tail(10) #Cek data
```

```python
# Identifikasi pelanggan baru berbasis level (Persona)
agg_df["customers_level_based"] = [f"{i[0]}_{i[1]}_{i[2]}_{i[-1]}" for i in agg_df.values]
```

```python
agg_df = agg_df.loc[:, ["customers_level_based", "PRICE"]].groupby("customers_level_based").agg({"PRICE": "mean"}).sort_values(by="PRICE", ascending=False ).reset_index()
agg_df["customers_level_based"].head() #Cek data lagi
```

```python
# Segmentasikan pelanggan baru (Persona)

agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], 4, labels=["D", "C", "B", "A"])
agg_df.head()
```

```python
# Jelaskan segmen dan terutama "C"

agg_df.groupby(["SEGMENT"]).agg({"PRICE": ["mean", "max", "sum"]})
agg_df[agg_df["SEGMENT"] == "C"].describe()
```
