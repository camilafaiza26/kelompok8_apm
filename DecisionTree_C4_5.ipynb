```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('bank.csv')
dataset
```

```python
#mengecek dataset apakah ada yang kosong
dataset.isnull().sum()
```

```python
#mengecek tipe data dari dataset
dataset.info()
```

```python
#untuk mengubah data string menjadi int
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
for column in dataset:
  if dataset[column].dtypes == object:
   dataset[column] = enc.fit_transform(dataset[column])
```

```python
dataset.info()
```

```python
dataset
```

```python
x = dataset.drop(['loan'], axis=1)
y = dataset['loan']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=500, max_depth=None, min_samples_split=2, min_samples_leaf=1,
                                    min_weight_fraction_leaf=0, max_leaf_nodes=None, min_impurity_decrease=0)
```

```python
clf = model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz

tree_data = export_graphviz(model, out_file=None, feature_names=x.columns, 
                                class_names = ['No', 'Yes'], filled= True)
```

```python
graph = graphviz.Source(tree_data)
graph.view()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(y_test, y_pred)
print("Tingkat Akurasi Algoritma C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(y_test, y_pred)
print('Tingkat Akurasi : {:.2%}'.format(akurasi))
```
